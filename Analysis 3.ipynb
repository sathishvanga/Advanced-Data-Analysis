### 2) What is the Average price of vehicle by fuel type and gearbox type.Give a plot?

```python
gdata = df.groupby(['fuelType', 'gearbox'])['price'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='fuelType', y='price', hue='gearbox', data=gdata)
plt.title('Average Price Based on Fuel Type and Gearbox')
plt.xlabel('Fuel Type')
plt.ylabel('Average Price')
plt.show()
```

The average price of vehicles varies based on fuel type and gearbox type, with hybrid fuel type having the highest average price, followed by diesel, and the lowest being for vehicles with other fuel types.

### 3)What is the Average power of a vehicle by vehicle type and gearbox type.Give a plot?

```python
data = df.groupby(by=['vehicleType','gearbox'])['powerPS'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='vehicleType', y='powerPS', hue='gearbox', data=data)
plt.title('Average Power Based on Vehicle Type and Gearbox')
plt.xlabel('Vehicle Type')
plt.ylabel('Average Power')
plt.show()
```

The average power of vehicles varies based on their type and gearbox type, with coupe vehicles having the highest average power, followed by SUVs, and kleinwagens having the lowest average power.

###  4) What is the Average price of a vehicle by brand as well as vehicle type.Use heatmap to explain this ?



```python
avg_price = df.groupby(['brand', 'vehicleType'])['price'].mean().reset_index()

heatmap_data = avg_price.pivot(index='brand', columns='vehicleType', values='price')

plt.figure(figsize=(15, 10))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt=".0f", cbar_kws={'label': 'Average Price'})
plt.title('Average Price of Vehicles by Brand and Vehicle Type')
plt.xlabel('Vehicle Type')
plt.ylabel('Brand')
plt.show()
```
